```python
import pandas as pd


def convert_csv_format(input_path: str, output_path: str) -> None:
    # Read source file
    df = pd.read_csv(input_path)

    # Map source columns to target format
    out = pd.DataFrame({
        "timestamp": pd.to_datetime(df["Date"], errors="coerce"),
        "Open": pd.to_numeric(df["Open"], errors="coerce"),
        "High": pd.to_numeric(df["High"], errors="coerce"),
        "Low": pd.to_numeric(df["Low"], errors="coerce"),
        "Close": pd.to_numeric(df["Price"], errors="coerce"),
        #Volume": 0
    })

    # Drop bad rows, sort oldest to newest, format date only
    out = out.dropna(subset=["timestamp", "Open", "High", "Low", "Close"])
    out = out.sort_values("timestamp").reset_index(drop=True)
    out["timestamp"] = out["timestamp"].dt.strftime("%Y-%m-%d")

    # Save
    out.to_csv(output_path, index=False)
    print(f"Saved converted file to: {output_path}")


if __name__ == "__main__":
    input_file = "US2Y1.csv"
    output_file = "US2Y.csv"
    convert_csv_format(input_file, output_file)
```
